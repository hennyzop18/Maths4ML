```python
import numpy as np

x = []
for i in range(10):
  x.append(np.random.randint(-61, 61))
print(x)
```

```python
x = np.sort(x)
print(x)
```

```python
import matplotlib.pyplot as plt
X = 2*x + 1
Y = x*x
plt.plot(x, X, 'r', label="2*x + 1")
plt.plot(x, Y, 'b', label="x*x")
plt.legend()
plt.show()
```
